# Logbooks in PHYS 409

This python notebook is an example of the expectations for experimental logbooks in PHYS 409. 

The logbook should reflect your learning process in the lab on a day-to-day basis:

- To keep track of your experimental process: Formulation of a question, hypothesis, prediction, testing and analysis
- To enable replication your experiment and, hopefully, your experimental results by another scientist or your future self 
- To enable sharing/reviewing (and marking) of your work

It should include theory and equations, details of apparatus, explicit and quantitative work plan, data and plots, issues encountered, conclusions drawn from measurements, and discussion on errors and uncertainties.

- Always date your logbook entries. There is no going back in time, there are only more ideas to discuss.
- Always record what you think and why.
- Always record later if it was right or not, and why.

Experimental physics is not textbook physics: uncertainties and discussions of uncertainties are expected. All of your analysis, including simulations and error analysis of experimental data, should happen in python, in your logbook.

## A good example of using the logbook

### September 5th, 2020

Here's a good example of using the logbook to document the development of an issue with the experimental data. This presents an opportunity to discuss the sources of noise in the experimental data, and how it might be improved.

```python
data = np.loadtxt('data.csv', delimiter=',', unpack=True)
plt.plot(data[0,:], data[1,:]);
plt.xlabel("x [arb.]");
plt.ylabel("y [arb.]");
```

Hmm, this looks like it ought to be sinusoidal. *checks physics*, yes, it ought to be. We are seeing a lot of noise in the data. I think the noise could be there because my lab partner was violently shaking the experimental table when we we're taking data. I've told them to stop that.

### September 6th, 2020

Today I snuck into the lab and took more data when my partner wasn't around.

```python
new_data = np.loadtxt('data_2.csv', delimiter=',', unpack=True)
plt.plot(data[0,:], data[1,:], label ='shaking');
plt.plot(new_data[0,:], new_data[1,:], label = 'no shaking');
plt.xlabel("x [arb.]");
plt.ylabel("y [arb.]");
plt.legend();
```

We see that the noise is way down! Our suscpicion was correct, gotta find a new lab partner.

## A bad example of using the logbook

This is an example of using the logbook to record a coding error. There's no physics to discuss here, and this sort of recoring should be avoided.

```python
plt.plot(data[0,:], data[1,:], label ='shaking');
plt.plot(new_data[1,:], new_data[0,:], label = 'no shaking');
plt.title('A title')
plt.xlabel("x [arb.]");
plt.ylabel("y [arb.]");
plt.legend();
```

Oh no! What's happening here, it's a mistake I've made so I'm recording it in my logbook. Trying again:

```python
plt.plot(data[0,:], data[1,:], label ='shaking');
plt.plot(new_data[0,:], new_data[1,:], label = 'no shaking');
plt.title('A title')
plt.xlabel("x [arb.]");
plt.ylabel("y [arb.]");
plt.legend();
```

There we go. Problem solved.
